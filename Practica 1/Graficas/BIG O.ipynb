```python
data = data.melt(id_vars=["model", "muestra"], 
        var_name="language", 
        value_name="time")
```

```python
data_2 = data.groupby(['model','muestra','language']).agg(
    {
        #'language' :'mean',
        'time' :['mean','std'] 
    }
).reset_index()
data_2.columns = ['model','muestra','language','mean','std']
data_2.reindex(columns=sorted(data_2.columns))
data_2
```

```python
mean = alt.Chart(data_2).mark_line().encode(
    alt.X('muestra:O', scale=alt.Scale(zero=False)),
    alt.Y('mean:Q',title='seconds MEAN'),
    alt.Color('model'),
    alt.Row('language'),
    tooltip = alt.Tooltip(['mean:Q','muestra:Q'])
).transform_filter(
    alt.FieldOneOfPredicate(field='model', oneOf=['Radix_sort','Merge_sort','Quick_sort','Insertion_sort'])
)

std = alt.Chart(data_2).mark_line().encode(
    alt.X('muestra:O'),
    alt.Y('std:Q',title='standard desviation'),
    alt.Color('model'),
    alt.Row('language'),
    tooltip = alt.Tooltip(['mean:Q','muestra:Q'])
).transform_filter(
    alt.FieldOneOfPredicate(field='model', oneOf=['Radix_sort','Merge_sort','Quick_sort','Insertion_sort'])
)

mean|std
```

```python
alt.Chart(data_2, title='Desviación Estandar vs Tamaño de lista').mark_line().encode(
    alt.X('muestra:O', title='Tamaño de lista'),
    alt.Y('std:Q',title='Desviación estandar'),
    alt.Color('language'),
    alt.Column('model', title='Algoritmo'),
    tooltip = alt.Tooltip(['mean:Q','muestra:Q']),
    #detail='model'
).transform_filter(
    alt.FieldOneOfPredicate(field='model', oneOf=['Radix_sort','Merge_sort','Quick_sort']),#'Insertion_sort'])
    #alt.FieldOneOfPredicate(field='language', oneOf=['Go']),
).properties(
    width=250,
    height=200
)
```
